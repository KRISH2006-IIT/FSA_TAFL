## 3. ***Data Wrangling***

### Data Wrangling Code

### What all manipulations have you done and insights you found?

Manipulations Performed:


*   Created a copy of the original dataset.
*   Removed irrelevant columns and duplicate records.
*   Dropped the order_date_time column due to excessive missing values.
*   Converted date columns to datetime format.
*   Removed records with inconsistent timestamps.
*   Checked remaining missing values and prepared a cleaned dataset for further analysis.

Insights:



*   The dataset contained duplicate and logically inconsistent records, which were removed.

*   order_date_time had a high percentage of missing values and was excluded.

*   Date columns were successfully converted for time-based analysis.
*   The cleaned dataset is now ready for Exploratory Data Analysis (EDA) and preprocessing.














````python
# Write your code to make your dataset analysis ready.
# Create a copy of the original dataset
df_copy = df.copy()


# Remove irrelevant columns
columns_to_drop = [
    'Unique id',
    'Order_id',
    'Customer Remarks',
    'Customer_City',
    'connected_handling_time'
]

df_copy.drop(columns=columns_to_drop, inplace=True)


# Check duplicate records
duplicate_count = df_copy.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)


# Remove duplicate records
df_copy.drop_duplicates(inplace=True)


# Verify duplicate removal
print("Remaining duplicate rows:", df_copy.duplicated().sum())
print("Dataset Shape:", df_copy.shape)


# Remove column with excessive missing values
df_copy.drop(columns=['order_date_time'], inplace=True)


# Convert date columns to datetime format
df_copy['Issue_reported at'] = pd.to_datetime(
    df_copy['Issue_reported at'],
    dayfirst=True
)

df_copy['issue_responded'] = pd.to_datetime(
    df_copy['issue_responded'],
    dayfirst=True
)

df_copy['Survey_response_Date'] = pd.to_datetime(
    df_copy['Survey_response_Date'],
    format='%d-%b-%y'
)


# Check for inconsistent timestamps
invalid_records = df_copy[
    df_copy['issue_responded'] < df_copy['Issue_reported at']
]

print("Number of inconsistent records:", len(invalid_records))


# Remove inconsistent records
df_copy = df_copy[
    df_copy['issue_responded'] >= df_copy['Issue_reported at']
].copy()


# Verify timestamp consistency
print("Remaining inconsistent records:",
      (df_copy['issue_responded'] < df_copy['Issue_reported at']).sum())

print("Final Dataset Shape:", df_copy.shape)


# Check remaining missing values
missing_values = pd.DataFrame({
    'Missing Values': df_copy.isnull().sum(),
    'Percentage (%)': round((df_copy.isnull().sum() / len(df_copy)) * 100, 2)
})

missing_values = missing_values[missing_values['Missing Values'] > 0]
missing_values


# Save the cleaned dataset
df_clean = df_copy.copy()
````
